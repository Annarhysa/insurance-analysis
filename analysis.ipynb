```python
#
df['claim date'] = pd.to_datetime(df['claim date'])
df['claim date'].dtype
df.head()
```

```python
# Function to calculate months until today
def months_until_today(original_date):
    # Get today's date
    today = datetime.now().date()
    
    # Calculate the number of months between original_date and today
    months = (today.year - original_date.year) * 12 + today.month - original_date.month
    
    return months
```

```python

df['claim_period'] = df['claim date'].apply(months_until_today)
df.head()
```
